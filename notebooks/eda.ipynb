# EDA - Insurance Medical Charges Dataset

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/insurance.csv')
df.head()
```

```
df.info()
```

```
df.describe(include='all')
```

```
df.isna().sum()
```

```
plt.figure(); plt.hist(df['charges'], bins=40); plt.xlabel('charges'); plt.ylabel('count'); plt.title('Distribution of charges'); plt.show()
```

```
plt.figure();
box_data=[df.loc[df['smoker']=='no','charges'], df.loc[df['smoker']=='yes','charges']]
plt.boxplot(box_data, labels=['no','yes']); plt.xlabel('smoker'); plt.ylabel('charges'); plt.title('Charges by smoker'); plt.show()
```

```
plt.figure();
for label in ['no','yes']:
    sub=df[df['smoker']==label]
    plt.scatter(sub['bmi'], sub['charges'], label=label, s=10, alpha=0.6)
plt.xlabel('bmi'); plt.ylabel('charges'); plt.title('BMI vs charges (by smoker)'); plt.legend(title='smoker'); plt.show()
```
